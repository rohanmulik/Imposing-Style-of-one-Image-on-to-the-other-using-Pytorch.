```python
#Import all needed Libraries


import torch
import torch.optim as optim
import numpy as np
from torchvision import transforms, models
from PIL import Image
from matplotlib import pyplot as plt

```

```python
vgg = models.vgg19(pretrained = True).features
for param in vgg.parameters():
    param.requires_grad_(False)
```

```python
img_path_content = '\Images\Lambo.jpg'
img_path_style = '\Images\c-rohan-kelly-storm-front-on-bondi-beach.jpg'
```

```python
content_image = Image.open(img_path_content).convert('RGB')

style_image = Image.open(img_path_style).convert('RGB')
```

```python
max_size = 400
shape = None
if max(content_image.size  ) > max_size:
    size = max_size
    
else:
    size = max(content_image.size )
    
if shape is not None:
    size = shape
    
in_transform = transforms.Compose([
                        transforms.Resize(size),
                        transforms.ToTensor(),
                        transforms.Normalize((0.485, 0.456, 0.406), 
                                             (0.229, 0.224, 0.225))])
content_image = in_transform(content_image)[:3,:,:].unsqueeze(0)

if max(style_image.size  ) > max_size:
    size = max_size
    
else:
    size = max(style_image.size )
    
if shape is not None:
    size = shape
style_image = in_transform(style_image)[:3,:,:].unsqueeze(0)
    
```

```python
def im_convert(tensor):
    """ Display a tensor as an image. """
    
    image = tensor.to("cpu").clone().detach()
    image = image.numpy().squeeze()
    image = image.transpose(1,2,0)
    image = image * np.array((0.229, 0.224, 0.225)) + np.array((0.485, 0.456, 0.406))
    image = image.clip(0, 1)

    return image
```

```python
# display the images
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
# content and style ims side-by-side
ax1.imshow(im_convert(content_image))
ax2.imshow(im_convert(style_image))
```
